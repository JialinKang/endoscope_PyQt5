```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_bgr = cv2.imread('./images/endo.png')
img_hsv = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2HSV)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
hsvChannels = cv2.split(img_hsv)
H = hsvChannels[0]
S = hsvChannels[1]
V = hsvChannels[2]
print(max(H.flatten()), max(S.flatten()), max(V.flatten()))
```

```python
V_mean = np.mean(V)
V_std = np.std(V)
t_v = V_mean + V_std * 0
S_mean = np.mean(S)
S_std = np.std(S)
t_s = S_mean - S_std * 1
width, heigh = H.shape
print(t_v, t_s)
```

```python
lower = np.array([0, 0, t_v])
upper = np.array([360,t_s,255])
 
mask = cv2.inRange(img_hsv, lower, upper)
res = cv2.bitwise_and(img_bgr, img_bgr, mask=mask)
plt.imshow(res)
```

```python
mask_rgb = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
plt.imshow(mask_rgb)
```

```python
ret,thresh1 = cv2.threshold(mask,127,255,cv2.THRESH_BINARY)
```

```python
print(thresh1.shape)
```

```python
plt.imshow(thresh1, 'gray')
```
