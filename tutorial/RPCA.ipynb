```python
import math
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
N = 100
num_groups = 3
num_values_per_group = 40
p_missing = 0.2

Ds = []
for k in range(num_groups):
    d = np.ones((N, num_values_per_group)) * (k + 1) * 10
    Ds.append(d)
    
D = np.hstack(Ds)
```

```python
D_img = Image.fromarray(D.astype(np.uint8))
plt.imshow(D_img)
```

```python
n1, n2 = D.shape
S = np.random.rand(n1,n2)
D[S < 0.2] = 0

D_nos = Image.fromarray(D.astype(np.uint8))
plt.imshow(D_nos)
```

```python
U,s,V = numpy.linalg.svd(D_nos, full_matrices=False)
```
